```python
import preprocessor as p
```

```python
train_df = pd.read_csv('/kaggle/input/tweets_202001121218.csv')
```

```python
train_df.head()
```

```python
train_df.count()
```

```python
train_df = train_df.dropna()
train_df = train_df.drop_duplicates()
```

```python
train_df.count()
```

```python
train_df.head()
```

```python
def preprocess_tweet(row):
    text = row['text']
    text = p.clean(text)
    return text
```

```python
train_df['text'] = train_df.apply(preprocess_tweet, axis=1)
```

```python
train_df.head()
```

```python
from gensim.parsing.preprocessing import remove_stopwords
```

```python
def stopword_removal(row):
    text = row['text']
    text = remove_stopwords(text)
    return text
```

```python
train_df['text'] = train_df.apply(stopword_removal, axis=1)
```

```python
train_df.head()
```

```python
train_df['text'] = train_df['text'].str.lower().str.replace('[^\w\s]',' ').str.replace('\s\s+', ' ')
```
